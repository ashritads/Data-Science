```python
#Quiz-1 Question 2
#BikeSharing Problem

import sklearn
import graphviz
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import sklearn.tree
import numpy as np
import urllib
import csv
import pandas as pd
from StringIO import StringIO
from sklearn.cross_validation import KFold
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
```

```python
### Initiate classifier
def scoreNaiveBayes(data, clf):
    X_matrix = data.iloc[:,1:11] 
    Y = data.iloc[:,0]

    from sklearn.cross_validation import KFold, cross_val_score
    
    kf = KFold(len(Y), n_folds=5, shuffle=True, random_state=0)
    scores = cross_val_score(clf, X_matrix, Y, cv=kf, scoring="f1")
    print "F1: {0:.3f}".format(np.mean(scores))
    scores = cross_val_score(clf, X_matrix, Y, cv=kf, scoring="precision")
    print "Precision: {0:.3f}".format(np.mean(scores))
    scores = cross_val_score(clf, X_matrix, Y, cv=kf, scoring="recall")
    print "Recall: {0:.3f}".format(np.mean(scores))
```

```python
print "Naïve Bayes classifier & logistic regression for Breast Cancer Wisconsin Dataset"

#read data
data = pd.read_csv(open("C:\Users\Ashrita\Documents\Machine Learning\Data Files\data.csv"), header=0)

#just checking data is loaded
print(data.shape)

# put the original column names in a python list
original_headers = list(data.columns.values)
```

```python
data = data.replace('M', 1)
data = data.replace('B', 0)
data = data.drop('id', axis = 1)
data.head()

#data_x = data.iloc[:,1:11]
#data_y = data.iloc[:,0]
```

```python
#from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

print "Gaussian Naive Bayes"
nb = GaussianNB()   # Features are continuous

scoreNaiveBayes(data,nb)
```

```python
from sklearn.linear_model import LogisticRegression

print "Logistic Regression"
lr = LogisticRegression()

scoreNaiveBayes(data,lr)
```

Naive Bayes Classifier is performing better than the logisitic regression. Logistic regression is a discriminative models, it uses only p(y|x). Naive Bayes is a generative model, it uses p(x,y). When we work with small datasets, a generative model can beat a discriminative model.
